```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
file_name = 'names'
path_file = ''.join((os.getcwd(), '\\', file_name, '\\'))
list_file_name = os.listdir(path=path_file)
```

```python
list_file_name = {}
for year in range(1900, 2001):
    list_file_name[year] = pd.read_csv(
        f'names/yob{year}.txt',
        names=['Name', 'Gender', 'Quantity'])
names_all = pd.concat(list_file_name).reset_index(level=1, drop=True)
```

```python
quantity_name = names_all[names_all['Name'].isin(['Ruth', 'Robert'])]
quantity_name = quantity_name.groupby([quantity_name.index, 'Name']).sum()
quantity_name.unstack('Name').plot()
```

```python
quantity_name = names_all[names_all['Name'].isin(['Ruth', 'Robert'])]
quantity_name = quantity_name[quantity_name.index.get_level_values(0).isin(range(1900, 2000, 5))]
quantity_name = quantity_name.groupby([quantity_name.index, 'Name']).sum()
quantity_name.plot(kind='hist', bins=10)
```

```python
quantity_name = names_all[names_all['Name'].str[0] == 'R']
quantity_name = quantity_name[quantity_name.index == 1950]
quantity_name = quantity_name.groupby(['Name']).sum().sort_values('Quantity', ascending=False).head(10)
quantity_name.plot(kind='pie', y='Quantity')
```

```python
quantity_name = names_all.groupby(['Name']).sum()
quantity_name['Letter'] = sum(quantity_name.index.str.count(letter) for letter in 'eyuioaEYUIOA')
quantity_name.plot(kind='scatter', y='Letter', x='Quantity', c='pink')
```
